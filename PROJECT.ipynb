### Background Information

The American Bible Society (ABS) was founded in 1816 with the aim of making the Bible affordable and accessible throughout the United States and the world. The organisation combined religious, educational, and charitable purposes, striving to ensure that as many people as possible, regardless of their social background or geographical location, could read the Bible. Over time, the ABS grew into one of the world's largest Bible publishers, printing and distributing billions of Bibles in multiple languages and formats. During the Civil War, ABS produced the first pocket Bible for soldiers, bringing the light of the gospel to the front lines of the nation's darkest periods. Furthermore, ABS has contributed to a wide range of developments in literacy and accessibility. In 1835, they translated the world's first Braille Bible, enabling blind people to read the Bible for the first time (American Bible Society, 2026).

“The new testament of our lord and saviour Jesus Christ: translated out of the original Greek & with the former translations diligently compared & revised” (The new testament of our lord and saviour Jesus Christ) is an edition of the New Testament published in New York in 1850 by the American Bible Society. The text is a translation from the Greek into English, dating from the 17th century. The phrase ‘with the former translations diligently compared & revised’ indicates that the translators placed great importance on the accuracy of the text and its continuity with earlier Bible translations. This edition is of particular significance for its materials and accessibility features. Unlike the standard printed Bibles of the time, this Bible was printed using raised lettering, allowing those who were blind or partially sighted to read it by touch. Rather than the Braille developed by Louis Braille, it adopted the Boston Line Type system, invented in 1835 by Samuel Gridley Howe, the first headmaster of the Perkins School for the Blind, since Louis's Braille had not been invented and announced.

Boston Line Type adopted a simplified Roman alphabet with angular, triangular shapes, based on the principle that straight lines and acute angles are easier to distinguish by touch than curves. Furthermore, to improve tactile readability, capital letters were omitted. During the 19th century, the Boston Line Type became one of the most widely used tactile reading systems in the United States and remained in widespread use for several decades until Braille became widely adopted. The manufacturing process for embossed books was not only technically demanding but also costly. Printing required assembling and securing metal plates moulded into the shapes of raised characters, which were then pressed onto thick paper. A close inspection reveals faint rectangular marks around the characters, traces of this mechanical printing process. The high cost of specialised printing equipment was likely one reason why embossed tactile books remained relatively limited in circulation, despite their importance in terms of accessibility.

One reason why linear embossing systems such as the Boston Line Type gained popularity was that they were designed to be readable not only by blind individuals but also by sighted people. This reflects the 19th-century view that tactile reading systems should not be designed solely with the needs of blind people in mind, but should also be visually recognisable to sighted people (Old Starbridge Village, 2024). Consequently, this New Testament is an important historical example of early accessible publishing practices, demonstrating how religious texts contributed to the development of inclusive reading technologies prior to the standardisation of Braille.

## Objectives

The aim of this project is to investigate, through an analysis of "The new testament of our lord and saviour Jesus Christ", the extent to which modern generative AI and image recognition technologies can interpret and preserve historically accessible reading systems. This edition is historically significant not only as a religious text but also as a testament to accessible 19th century publishing. This is because it was produced using Boston Line Type, an embossed typeface that served as a tactile reading system developed prior to the widespread adoption of Braille. In particular, this project will verify whether AI systems can accurately recognise and interpret the embossed characters of the Boston Line Type from digital images of the text, and convert the recognised content into modern Braille. In doing so, we will explore the relationship between historical tactile communication systems and modern machine interpretation technologies, whilst examining how accessibility practices have evolved from embossed printing to the interpretation of digital formats.

Furthermore, this project aims to critically examine the collaborative role of humans and AI in the interpretation and translation process. Rather than positioning AI as an autonomous interpretation tool, we conceptualise it as a computational tool that supports human-led archival interpretation through pattern recognition, image analysis and text generation. Human involvement is indispensable in the selection of materials, historical contextualisation, the verification of AI-generated outputs, and the correction of inaccuracies in transcription and translation. By evaluating both the capabilities and limitations of AI systems when processing historical tactile texts, this project aims to assess whether AI can make a meaningful contribution to the preservation and reinterpretation of historical accessibility technologies within cultural heritage collections.
